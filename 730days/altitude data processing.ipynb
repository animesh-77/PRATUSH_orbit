```python
x_vals= np.array(df["lunar_probe.ElapsedDays"])
y_vals= np.array(df["lunar_probe.Luna.Altitude"])

plt.figure(1)

plt.plot(x_vals, y_vals)

plt.xlim(0, 730)

plt.title("Altitude of s/c for 730 days", fontsize=15) #  SET according to file opened
plt.xlabel("Days Elapsed", fontsize=15)
plt.ylabel("Altitude in km", fontsize=15)
plt.grid()
plt.show()
```

```python
# plotting altitude of Apolune and Perilune only


df1= df[(abs(df["lunar_probe.Luna.TA"])> 175) & (abs(df["lunar_probe.Luna.TA"])< 185)]# dataset of values at apolune  only
df2= df[(abs(df["lunar_probe.Luna.TA"])<5) | (abs(df["lunar_probe.Luna.TA"])>355)]# dataset of values at perilune only


df2.head()

x_vals_a= np.array(df1["lunar_probe.ElapsedDays"])
y_vals_a= np.array(df1["lunar_probe.Luna.Altitude"])

x_vals_p= np.array(df2["lunar_probe.ElapsedDays"])
y_vals_p= np.array(df2["lunar_probe.Luna.Altitude"])

plt.figure(2)

plt.plot(x_vals_a, y_vals_a, label= "Apolune", color="Orange")
plt.plot(x_vals_p, y_vals_p, label= "Perilune" ,color= "Blue")

plt.xlim(0, 730)

plt.title("Altitude of Apolune and Perilune for 730 days", fontsize=15) #  SET according to file opened
plt.xlabel("Days Elapsed", fontsize=15)
plt.ylabel("Altitude in km", fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

```python
# plotting longitude of perilune (=RAAN +AOP) only when s/c is at Apolune or Perilune

df3=pd.concat([df1,df2])
df3=df3.sort_index()


y_vals= np.array(df3["lunar_probe.LunaFixed.RAAN"]+ df3["lunar_probe.LunaFixed.AOP"])

y_vals= np.where(y_vals>360, y_vals-360, y_vals)

x_vals= np.array(df3["lunar_probe.ElapsedDays"])

plt.figure(3)
plt.plot(x_vals, y_vals, label="Longitude of Perilune")
plt.axhline(y=180, color= "orange", label="180")
plt.xlim(0, 730)
plt.ylim(0, 360)

plt.title("Longitude of Perilune for 730 days", fontsize=15) #  SET according to file opened
plt.xlabel("Days Elapsed", fontsize=15)
plt.ylabel("Longitude in LunaFixed", fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

# Getting rid of the horizontal lines when plotting altitude vs longitude in LunaFixed system

### when longitude goes from 180 to -170 or 170 to -180 a horizontal line is added. The idea is to take small arrays of x-y plottable data from dataframe in which longitude value is in strictly increasing order. Corresponding altitude values are the plottable y-values. These x-y values are plotted one array at a time while traversing through full dataset.

```python
x_vals= np.array(df["lunar_probe.Luna.Longitude"])
y_vals= np.array(df["lunar_probe.Luna.Altitude"])

plot_x=np.zeros(0)  # to store plottable values during each orbit... bascially x vals should be in insreasing order
plot_y=np.zeros(0)

plt.figure(4)

for i, long_val in enumerate(x_vals):
    if (abs(long_val)> 175 ):  # to skip horizontal lines when going 170 to -180 or 180 to -170 
        
            
            if (long_val<0): # long value is -ve
                    
                plot_x= np.append(plot_x, -x_vals[i])# add altitude to the positive side
                plot_y= np.append(plot_y, y_vals[i]) 
                    
                plt.plot(plot_x, plot_y, linewidth=.01 ,color="red")     # plot plottable values
                
                plot_x=np.zeros(0)   # dump the array values
                plot_y=np.zeros(0)
                    
                plot_x= np.append(plot_x, x_vals[i])# to the new array add the most negative number
                plot_y= np.append(plot_y, y_vals[i]) 
                    
            else:
                plot_x= np.append(plot_x, x_vals[i])# add a last postitve value
                plot_y= np.append(plot_y, y_vals[i])
                
                plt.plot(plot_x, plot_y, linewidth=.01 ,color="red")     # plot plottable values
                
                plot_x=np.zeros(0)   # dump the array values
                plot_y=np.zeros(0)
                    
                plot_x= np.append(plot_x, -x_vals[i])# to the new array add the most negative number
                plot_y= np.append(plot_y, y_vals[i])
         
        
        
        
    else:
        plot_x= np.append(plot_x, x_vals[i])  # plottable x_values should be in increasing order
        plot_y= np.append(plot_y, y_vals[i])
 

    
#####################################################################


plt.xlim(-180, 180)
plt.xticks(np.arange(-180, 180, 10))
plt.title("Altitude of s/c for 730 days", fontsize=15) #  SET according to file opened
plt.xlabel("Longitude LunaFixed", fontsize=15)
plt.ylabel("Altitude", fontsize=15)
plt.grid()
plt.show()

########################################################################
```
