```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

import ipywidgets as wids
from ipywidgets import interact
```

```python
df=pd.read_csv("inc2.csv", header= 0)


df.columns
```

```python
df.iloc[:,0].unique()
```

```python
# inc= df.iloc[:,0].unique()
inc= np.zeros(0)
inc_val=0
max_day= np.zeros(0)
max_val=0



for index,row in df.iterrows():
    
    if index== len(df)-1:
        max_val= row["lunar_probe.ElapsedDays"]
        max_day= np.append(max_day, max_val)
        inc_val= np.round(row["inc"],2)
        inc= np.append(inc, inc_val)
        print(inc_val, " ", max_val)



    elif row["lunar_probe.ElapsedDays"] > df.iloc[index+1]["lunar_probe.ElapsedDays"]:
        
        max_val= row["lunar_probe.ElapsedDays"]
        max_day= np.append(max_day, max_val)
        inc_val= np.round(row["inc"],1)
        inc= np.append(inc, inc_val)
        print(inc_val, " ", max_val)

        

```

making a dataframe where first column is value of initial incliantion used in simualtion and other column is the number of days the spacecraft orbitted before falling below 20km. The maximum no. of days is 60 due to the nature of simulation

```python
df1= pd.DataFrame({"inc": inc, "days":max_day})
df1
```

```python
%matplotlib qt

fig= plt.figure(figsize= (12,10))


plt.style.use('ggplot')

plt.scatter( df1["inc"], df1["days"], s=90, color= "red")
plt.grid(True, color="black")
plt.xlabel("Inclination value", fontsize=25)
plt.ylabel("Number of days ", fontsize=25)

plt.xlim(-5,95)
plt.xticks(np.arange(0,95,5), fontsize= 15)
plt.yticks(fontsize= 15)

plt.show()
```
