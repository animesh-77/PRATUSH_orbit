```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("690days_near_equatorial.csv", header=0)
```

```python
display(df)
df.plot(subplots= True, figsize= (10,10))

```

```python
df.to_csv("690days_MJ2000.csv", index=None)
```

```python
df.min()
```

```python
alt_vals= np.array(df["chandrayaan.Luna.Altitude"])
day_vals= np.array(df["chandrayaan.ElapsedDays"])

plt.figure(1, figsize=(14,4), dpi=100)


plt.plot(day_vals, alt_vals)
plt.xlim(0,700)
plt.xlabel("Days Elapsed", fontsize=25)
plt.ylabel("Altitude in km", fontsize= 25)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.grid(True)

```

```python
%matplotlib inline
ecc_vals= np.array(df["chandrayaan.Luna.ECC"])
day_vals= np.array(df["chandrayaan.ElapsedDays"])

fig= plt.figure(2, figsize=(14,4), dpi=100)


plt.plot(day_vals, ecc_vals)
plt.xlabel("Days Elapsed", fontsize=25)
plt.ylabel("Eccentricity", fontsize=25)
plt.xlim(0,700)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.grid(True)
```

```python
sma_vals= np.array(df["chandrayaan.Luna.SMA"])
day_vals= np.array(df["chandrayaan.ElapsedDays"])

fig= plt.figure(3, figsize=(14,4), dpi=100)

plt.plot(day_vals, sma_vals)
plt.xlabel("Days Elapsed", fontsize=25)
plt.ylabel("SMA in km", fontsize=25)
plt.xlim(0,700)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.grid(True)
```

```python
%matplotlib inline
raan_vals= np.array(df["chandrayaan.Moon_MJ2000.RAAN"])
day_vals= np.array(df["chandrayaan.ElapsedDays"])

fig= plt.figure(4, figsize=(14,4), dpi=100)

plt.plot(day_vals, raan_vals)
plt.xlabel("Days Elapsed", fontsize=25)
plt.ylabel("RAAN", fontsize=25)
plt.xlim(0,700)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.grid(True)
```
